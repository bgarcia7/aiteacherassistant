```python
def save_example(prompt, response, notes=None, filename='examples.json'):
    data = json.loads(open(filename).read())
    if not data:
        data = []
    data.append({'prompt':prompt, 'response':response, 'notes':notes})
    file = open(filename, 'w')
    file.write(json.dumps(data))
    file.close()
```

```python
def get_jurisdictions():
    r = requests.get('http://commonstandardsproject.com/api/v1/jurisdictions/')
    data = json.loads(r.content)['data']
    return data
```

```python
def get_standard_sets(entity_id):
    r = requests.get('https://commonstandardsproject.com/api/v1/jurisdictions/{entity_id}'.format(entity_id=entity_id))
    data = json.loads(r.content)['data']
    return data
```

```python
def get_standard(standard_id="49FCDFBD2CF04033A9C347BFA0584DF0_D2604890_grade-01"):
    r = requests.get('https://api.commonstandardsproject.com/api/v1/standard_sets/{standard_id}'.format(standard_id=standard_id))
    data = json.loads(r.content)['data']
    return data
```

```python
def get_response(prompt):
    response = openai.Completion.create(
            model="text-davinci-003",
            prompt=prompt,
            temperature=0.6,
            max_tokens=3500 - len(prompt.split()),
            top_p=1,
            frequency_penalty=1,
            presence_penalty=1
    )
    return response.choices[0].text.strip()
```
